```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
july=[12,15,11,9,1,9,21]
august=[17,5,2,11,1,8,29]
plt.plot(x,july,linestyle='dashed', marker='D', color='r')
plt.plot(x,august,linestyle='dotted', marker='D', color='b')
plt.title('birthrate for july and august')
plt.ylabel('new births')
plt.xlabel('days')
plt.grid()
plt.show()
```
