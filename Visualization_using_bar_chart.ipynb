```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7]
july = [12, 15, 11, 9, 1, 9, 21]
august = [17, 5, 2, 11, 1, 8, 29]
plt.bar(x, july, color='r')
plt.title('Birthrate for July')
plt.ylabel('New Births')
plt.xlabel('Days')
plt.grid(axis='y')
plt.show()
```

```python
plt.bar(x, august, color='b')
plt.title('Birthrate for August')
plt.ylabel('New Births')
plt.xlabel('Days')
plt.grid(axis='y')
plt.show()
```
