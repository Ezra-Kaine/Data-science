```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
files = files.upload()
penguins_df = pd.read_csv("Penguins Data (1).csv")
penguins_df = penguins_df.dropna(subset=['Body Mass (g)'])
plt.hist(penguins_df['Body Mass (g)'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Penguin Body Mass')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
