### **Import Python libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Read the data and upload it to a dataframe**

```python
df=pd.read_csv('country_vaccinations.csv')
df.head(10)
```

### **Using .isnull() with .any()**

```python
df.isnull().any()
```

**Using .isnull() with .sum()**

```python
df.isnull().sum()
```

###**Visualize missing values**

```python
subset=df.iloc[:5200, :]
plt.figure(figsize=(12,8))
sns.heatmap(subset.isnull(), cbar=False, cmap='winter')
plt.show()
```

###**Dropping null values using .dropna()**

```python
df.dropna()
```

###**Using .fillna() to fix the null values**

```python
df.fillna(1)
```

###**Using Backward filling(.bfill())**

```python
df.fillna(method='bfill')
```

###**Using Forward filling(.ffill())**

```python
df.fillna(method='ffill')
```
