```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Penguins Data (3).csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

```python
x_col = 'Culmen Length (mm)'
y_col = 'Body Mass (g)'

sns.scatterplot(x=x_col, y=y_col, data=df, hue='Species', palette='viridis')
plt.title('Scatter Plot: Culmen Length vs Body Mass')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.tight_layout()
plt.show()
```

```python
x_col = 'Culmen Depth (mm)'
y_col = 'Body Mass (g)'

sns.scatterplot(x=x_col, y=y_col, data=df, hue='Species', palette='coolwarm')
plt.title('Scatter Plot: Culmen Depth vs Body Mass')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.tight_layout()
plt.show()
```

```python
df_sorted = df.sort_values(by='Culmen Length (mm)')

plt.fill_between(df_sorted['Culmen Length (mm)'], df_sorted['Body Mass (g)'], color='skyblue', alpha=0.6)
plt.plot(df_sorted['Culmen Length (mm)'], df_sorted['Body Mass (g)'], color='blue')  # Line over the area
plt.title('Area Plot: Culmen Length vs Body Mass')
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Body Mass (g)')
plt.tight_layout()
plt.show()
```
