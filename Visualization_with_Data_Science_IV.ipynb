```python
import matplotlib.pyplot as plt
blood_sugar_men=[80,83,100,112,129,135,150,140,89,99,113,120,105]
blood_sugar_women=[67,112,99,123,135,89,97,125,113,100,89,132,76]
type=[blood_sugar_men, blood_sugar_women]
color=['g','r']
label=['men','women']
bin=[60,80,100,120,140,160]
plt.xlabel('Blood sugar range')
plt.ylabel('Tatal number of patients')
plt.figure(figsize=(20,20))
plt.hist(type,color=color,label=label,bins=bin,rwidth=0.8,orientation='vertical',edgecolor='b')
plt.legend()
plt.show()
```
